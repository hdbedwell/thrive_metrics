#Number of students who have participated in an individual (1:1) mental health session delivered by an SRC up to and including the sixth week post the first preventative service. Need this for both crisis/preventative AND 1:1 / 1:2+

```python
#Find students served more than one time in quarter
#Source: https://stackoverflow.com/questions/14657241/how-do-i-get-a-list-of-all-the-duplicate-items-using-pandas-in-python
#stud_dups = pd.concat(g for _, g in q2.groupby("student_id") if len(g) > 1)
```

```python
#Find students served more than one time within six weeks 'Date of provided support'
# Source: https://stackoverflow.com/questions/54669337/pandas-dataframe-difference-between-all-dates-for-each-unique-id

stud_dups_sorted = q2.sort_values(['student_id', 'Date of provided support'])
```

```python
stud_dups_sorted['Datetime'] = pd.to_datetime(stud_dups_sorted['Date of provided support'])                              
g = stud_dups_sorted.groupby('student_id')['Datetime']
print(stud_dups_sorted.groupby('student_id')['Datetime'].apply(lambda g: g.iloc[0] - g.iloc[-1])) 
```

```python
### df['Time in Status'] = df.groupby('id')['Datetime'].diff(-1)*(-1)
#Time betwen service provides number of days between dates of service for each student

stud_dups_sorted['time_between_service'] = stud_dups_sorted.groupby('student_id')['Datetime'].diff(-1)*(-1)
stud_dups_sorted
```

```python
stud_dups_final = pd.read_excel(r"/Users/holly/Desktop/12292020 Thrive 6 wk Data.xlsx")
stud_dups_final
```

```python
type('time_between_service')
```

```python
#Find duplicate students supported where time_between_service is not null AND is less than 43 days. This will yield all students supported more than once within 6 weeks.
#Source: https://queirozf.com/entries/pandas-query-examples-sql-like-syntax-queries-in-dataframes

time_between_notnull = stud_dups_final.query('time_between_service != "Nan"')
time_between_notnull
```

```python
time_between_notnull_less6wks = time_between_notnull.query('time_between_service < 43.0')
time_between_notnull_less6wks
```

```python
#The number of students who have participated in an individual (1:1) mental health session delivered by an SRC up to and 
#including the sixth week post the first preventative service.
prevent_individual_6wks = time_between_notnull_less6wks.query('individual_group == "1:1 (individual)" & service_type == "Preventive (Ongoing)"').student_id.nunique()
prevent_individual_6wks
```

```python
#The number of students who have participated in an individual (1:2+) mental health session delivered by an SRC up to and 
#including the sixth week post the first preventative service.
prevent_group_6wks = time_between_notnull_less6wks.query('individual_group == "1:2+ (group)" & service_type == "Preventive (Ongoing)"').student_id.nunique()
prevent_group_6wks
```

```python
#The number of students who have participated in an individual (1:1) mental health session delivered by an SRC up to and 
#including the sixth week post the first crisis service.
crisis_group_6wks = time_between_notnull_less6wks.query('individual_group == "1:2+ (group)" & service_type == "Crisis"').student_id.nunique()
crisis_group_6wks
```

```python
#The number of students who have participated in an individual (1:1) mental health session delivered by an SRC up to and 
#including the sixth week post the first crisis service.
crisis_individual_6wks = time_between_notnull_less6wks.query('individual_group == "1:1 (individual)" & service_type == "Crisis"').student_id.nunique()
crisis_individual_6wks
```
